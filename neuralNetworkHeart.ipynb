```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import matplotlib.pyplot as plt


#Import Data and Setting up Target Vector
df = pd.read_csv("LungCancerDataset.csv")  

X_toScale = df.drop('PULMONARY_DISEASE', axis=1)
y = df['PULMONARY_DISEASE'].map({'NO': 0, 'YES': 1})
```

```python
#Scaling and Splitting Data, Defining Regularization Values
scaler = StandardScaler()
X = scaler.fit_transform(X_toScale)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

c = [0, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
```

```python
def runNNBase(RegValues): 
    print("Base NN") 
    for val in RegValues:
        print(f"Lambda: {val}")
        #Training Model
        nn = MLPClassifier(hidden_layer_sizes=(10), activation='logistic', alpha= val, max_iter=5000)
        nn.fit(X_train, y_train)
        
        #Cross Validation Score
        cv_scores = cross_val_score(nn, X_train, y_train)
        print(f"Mean Cross-Validation Accuracy: {cv_scores.mean()}")

        #Training Metrics
        y_train_pred = nn.predict(X_train)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_precision = precision_score(y_train, y_train_pred, zero_division=0)
        train_recall = recall_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred)

        #Test Metrics
        y_test_pred = nn.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        test_precision = precision_score(y_test, y_test_pred, zero_division=0)
        test_recall = recall_score(y_test, y_test_pred)
        test_f1 = f1_score(y_test, y_test_pred)

        print(f"Train Precision: {train_precision}")
        print(f"Train Recall: {train_recall}")
        print(f"Train F1-Score: {train_f1}")
        print(f"Train Accuracy {train_accuracy}")

        print(f"Test Precision: {test_precision}")
        print(f"Test Recall: {test_recall}")
        print(f"Test F1-Score: {test_f1}")
        print(f"Test Accuracy: {test_accuracy}")

        print()
```

```python
def runNNBasePlus(RegValues): 
    print("Base Plus NN") 
    for val in RegValues:
        print(f"Lambda: {val}")
        #Training Model
        nn = MLPClassifier(hidden_layer_sizes=(20,10), activation='logistic', alpha= val, max_iter=5000)
        nn.fit(X_train, y_train)
        
        #Cross Validation Score
        cv_scores = cross_val_score(nn, X_train, y_train)
        print(f"Mean Cross-Validation Accuracy: {cv_scores.mean()}")

        #Training Metrics
        y_train_pred = nn.predict(X_train)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_precision = precision_score(y_train, y_train_pred, zero_division=0)
        train_recall = recall_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred)
    
        #Test Metrics
        y_test_pred = nn.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        test_precision = precision_score(y_test, y_test_pred, zero_division=0)
        test_recall = recall_score(y_test, y_test_pred)
        test_f1 = f1_score(y_test, y_test_pred)

        print(f"Train Precision: {train_precision}")
        print(f"Train Recall: {train_recall}")
        print(f"Train F1-Score: {train_f1}")
        print(f"Train Accuracy {train_accuracy}")

        print(f"Test Precision: {test_precision}")
        print(f"Test Recall: {test_recall}")
        print(f"Test F1-Score: {test_f1}")
        print(f"Test Accuracy: {test_accuracy}")

        print()
```

```python
def runNNWide(RegValues): 
    print("Wide NN") 
    for val in RegValues:
        print(f"Lambda: {val}")
        #Training Model
        nn = MLPClassifier(hidden_layer_sizes=(75,75), activation='logistic', alpha= val, max_iter=5000)
        nn.fit(X_train, y_train)
        
        #Cross Validation Score
        cv_scores = cross_val_score(nn, X_train, y_train)
        print(f"Mean Cross-Validation Accuracy: {cv_scores.mean()}")

        #Training Metrics
        y_train_pred = nn.predict(X_train)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_precision = precision_score(y_train, y_train_pred, zero_division=0)
        train_recall = recall_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred)

        #Test Metrics
        y_test_pred = nn.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        test_precision = precision_score(y_test, y_test_pred, zero_division=0)
        test_recall = recall_score(y_test, y_test_pred)
        test_f1 = f1_score(y_test, y_test_pred)

        print(f"Train Precision: {train_precision}")
        print(f"Train Recall: {train_recall}")
        print(f"Train F1-Score: {train_f1}")
        print(f"Train Accuracy {train_accuracy}")

        print(f"Test Precision: {test_precision}")
        print(f"Test Recall: {test_recall}")
        print(f"Test F1-Score: {test_f1}")
        print(f"Test Accuracy: {test_accuracy}")

        print()
```

```python
def runNNDeep(RegValues): 
    print("DEEP NN") 
    for val in RegValues:
        print(f"Lambda: {val}")
        #Training Model
        nn = MLPClassifier(hidden_layer_sizes=(50,40,30), activation='logistic', alpha= val, max_iter=5000)
        nn.fit(X_train, y_train)
        
        #Cross Validation Score
        cv_scores = cross_val_score(nn, X_train, y_train)
        print(f"Mean Cross-Validation Accuracy: {cv_scores.mean()}")

        #Trining Metrics
        y_train_pred = nn.predict(X_train)
        train_accuracy = accuracy_score(y_train, y_train_pred)
        train_precision = precision_score(y_train, y_train_pred, zero_division=0)
        train_recall = recall_score(y_train, y_train_pred)
        train_f1 = f1_score(y_train, y_train_pred)
    
        #Test Metrics
        y_test_pred = nn.predict(X_test)
        test_accuracy = accuracy_score(y_test, y_test_pred)
        test_precision = precision_score(y_test, y_test_pred, zero_division=0)
        test_recall = recall_score(y_test, y_test_pred)
        test_f1 = f1_score(y_test, y_test_pred)

        print(f"Train Precision: {train_precision}")
        print(f"Train Recall: {train_recall}")
        print(f"Train F1-Score: {train_f1}")
        print(f"Train Accuracy {train_accuracy}")

        print(f"Test Precision: {test_precision}")
        print(f"Test Recall: {test_recall}")
        print(f"Test F1-Score: {test_f1}")
        print(f"Test Accuracy: {test_accuracy}")

        print()
```
